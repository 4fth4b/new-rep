```python
import pandas as pd
import os
import numpy as np
```

```python
df=pd.read_csv('HR_comma_sep (1).csv')
```

```python
df1=df.copy()
```

```python
df1
```

```python
a=pd.get_dummies(df1['Department'])
b=pd.get_dummies(df1['salary'])
```

```python
df1=pd.concat([df1,a,b],axis=1)
```

```python
df1
```

```python
df1.drop(['Department','salary'],axis=1,inplace=True)
```

```python
df1
```

```python
x=df1.drop('left',axis=1)
```

```python
y=df1['left']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
logistic=LogisticRegression()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=1)
```

```python
logistic.fit(xtrain,ytrain)
```

```python
logistic.score(xtrain,ytrain)
```
