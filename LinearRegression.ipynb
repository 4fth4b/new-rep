```python
import pandas as pd
import os
import numpy as np
```

```python
df=pd.read_csv('carprices (4).csv')
```

```python
df1=df.copy()
```

```python
df1
```

```python
df1.describe()
```

```python
a=pd.get_dummies(df1['Car Model'])
```

```python
df1=pd.concat([df1,a],axis=1)
```

```python
df1
```

```python
df1.drop('Car Model',axis=1,inplace=True)
```

```python
x=df1.drop(['Sell Price($)'],axis=1)
```

```python
y=df1['Sell Price($)']
```

```python
df1
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear=LinearRegression()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=.8,random_state=1)
```

```python
len(xtrain)
```

```python
linear.fit(xtrain,ytrain)
```

```python
linear.score(xtest,ytest)
```

```python
y_pred=linear.predict([[50000,6,0,1,0]])
```
